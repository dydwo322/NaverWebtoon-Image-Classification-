```python
# title_id=['703846','641253','748105','650305','670143','616239','671421','64997','318995','667573','675554']
title_id=['648419','131385','711422','728750','183559','685460','703839','750558','602910']
```

```python
import requests
from bs4 import BeautifulSoup as bs
from urllib.request import urlretrieve
 
from tqdm.notebook import tqdm
page=[]
tt=[]
au=[]
for i in tqdm(range(9)):

    url='https://comic.naver.com/webtoon/list?titleId={}'.format(title_id[i])
    resp=requests.get(url)
    soup=bs(resp.content,'lxml')
    # img_tags=soup.select('table.viewList tr td a img')
    tit_tags=soup.select('div.detail h2 span.title')
    auth_tags=soup.select('div.detail h2 span.wrt_nm')

    page_tag =soup.select("table.viewList tr td.title a")
    tt.append(tit_tags[0].text)
    au.append(auth_tags[0].text)   
    page.append(int(page_tag[0]["href"].split("&")[1].strip("no=")))
```

```python
au1=[]
for i in range(9):
  au1.append(au[i].strip())
```

```python
au1
```

```python
au2=[]
for i in au1:
  au2.append(i.replace(' ','').replace('/','_'))
```

```python
page
```

```python
lstpg=[]
for i in tqdm(range(9)):
    if page[i] %10 == 0:
        lstpg.append(page[i]//10)
    else:
        lstpg.append(page[i]//10+1)
```

```python
lstpg
```

```python
from tqdm.notebook import tqdm
dic_info={
    '제목':[],
    '작가':[],
    '이미지':[]
}

titleid=[]
author_name=[]

for i in tqdm(range(9)):
    for k in range(1,lstpg[i]+1):
      url='https://comic.naver.com/webtoon/list?titleId={}&page={}'.format(title_id[i],k)
      resp=requests.get(url)
      soup=bs(resp.content,'lxml')
      # img_tags=soup.select('table.viewList tr td a img')
      table= soup.select('table.viewList')
      thumbnail= table[0].find_all('img')

      
  

      for n in thumbnail:
        if 'thumbnail' in n['src']:
          dic_info['이미지'].append(n['src'])
          dic_info['작가'].append(au2[i])
          dic_info['제목'].append(tt[i])
          titleid.append(title_id[i])
          author_name.append(au2[i])


        
```

```python
pagenumbers=[]
for pg in range(len(dic_info['이미지'])):
  pagenumbers.append(int(dic_info['이미지'][pg].split('/')[-2]))
```
